# Feature Engineering

## ตรวจสอบข้อมูล
1. ดูข้อมูลที่เป็นประเภทตัวเลขและหมวดหมู่
2. ค้นหาค่า missing values
3. ดูค่า unique ของตัวแปรประเภทหมวดหมู่

## จัดการค่าหายไป (Missing Values)
1. เติมค่าหายไปในตัวเลขด้วยค่ามัธยฐานหรือค่าเฉลี่ย
2. เติมค่าหายไปในหมวดหมู่ด้วยค่าที่พบบ่อยที่สุด

## แปลงข้อมูลหมวดหมู่
1. ใช้ One-Hot Encoding กับค่าที่มีไม่เกิน 5-10 categories
2. ใช้ Label Encoding กับค่าที่มีหมวดหมู่จำนวนมาก
3. ใช้ Target Encoding หรือ Frequency Encoding สำหรับตัวแปรที่อาจสัมพันธ์กับ target

## สร้าง Feature ใหม่
1. การคำนวณค่าดัชนี เช่น BMI categories (ต่ำ, ปกติ, สูง)
2. การรวมค่าที่เกี่ยวข้อง เช่น การนับจำนวนปัจจัยเสี่ยงต่อโรคหัวใจ
3. การสร้าง interaction features เช่น BMI * Age

## Feature Selection
1. ใช้ PCA หรือ Feature Importance จาก RandomForest/XGBoost เพื่อลดมิติข้อมูล

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# โหลดข้อมูล
train_path = "/kaggle/input/hearth-disease-recognition/train.csv"
df = pd.read_csv(train_path)

# ลบค่าหายไปใน target
df = df.dropna(subset=["History of HeartDisease or Attack"])

# เติมค่าหายไปในค่าตัวเลขด้วยค่ามัธยฐาน
df["Body Mass Index"].fillna(df["Body Mass Index"].median(), inplace=True)

# เติมค่าหายไปในค่าหมวดหมู่ด้วยค่าที่พบบ่อยที่สุด
categorical_cols = ["Told High Cholesterol", "General Health", "Diagnosed Diabetes", "Difficulty Walking"]
for col in categorical_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)

# แปลง Yes/No เป็น 0/1
binary_cols = [
    "High Blood Pressure", "Told High Cholesterol", "Cholesterol Checked",
    "Smoked 100+ Cigarettes", "Diagnosed Stroke", "Diagnosed Diabetes",
    "Leisure Physical Activity", "Heavy Alcohol Consumption",
    "Health Care Coverage", "Doctor Visit Cost Barrier", "Difficulty Walking",
    "Vegetable or Fruit Intake (1+ per Day)"
]

for col in binary_cols:
    df[col] = df[col].map({"Yes": 1, "No": 0})

# แปลงค่าหมวดหมู่แบบลำดับ (Ordinal Encoding)
ordinal_mappings = {
    "General Health": {"Very Poor": 1, "Poor": 2, "Fair": 3, "Good": 4, "Excellent": 5},
    "Education Level": {"Never attended school": 1, "Some school": 2, "High school graduate": 3, 
                        "Some college": 4, "College graduate": 5},
    "Income Level": {"Less than $10,000": 1, "$10,000 - $14,999": 2, "$15,000 - $24,999": 3, 
                     "$25,000 - $34,999": 4, "$35,000 - $49,999": 5, "$50,000 - $74,999": 6, 
                     "$75,000 or more": 7}
}

for col, mapping in ordinal_mappings.items():
    df[col] = df[col].map(mapping)

# แปลงค่าหมวดหมู่ของ Sex เป็นตัวเลข
df["Sex"] = df["Sex"].map({"Male": 1, "Female": 0})

# เติมค่า NaN ที่ยังคงมีอยู่ด้วยค่ากลางของแต่ละคอลัมน์
for col in ordinal_mappings.keys():
    df[col].fillna(df[col].median(), inplace=True)

# สร้าง BMI Category
def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"

df["BMI Category"] = df["Body Mass Index"].apply(categorize_bmi)
df["BMI Category"] = df["BMI Category"].map({"Underweight": 1, "Normal": 2, "Overweight": 3, "Obese": 4})

# สร้าง Risk Score (นับจำนวนปัจจัยเสี่ยง)
risk_factors = ["High Blood Pressure", "Told High Cholesterol", "Smoked 100+ Cigarettes", "Diagnosed Stroke",
                "Diagnosed Diabetes", "Heavy Alcohol Consumption", "Difficulty Walking"]
df["Total Risk Score"] = df[risk_factors].sum(axis=1)

# สร้าง Age Group
def categorize_age(age):
    if age < 40:
        return "Young"
    elif age < 60:
        return "Middle-Aged"
    else:
        return "Senior"

df["Age Group"] = df["Age"].apply(categorize_age)
df["Age Group"] = df["Age Group"].map({"Young": 1, "Middle-Aged": 2, "Senior": 3})

# ลบ ID (ไม่มีผลกับการพยากรณ์)
df.drop(columns=["ID"], inplace=True)

# เติมค่าที่พบบ่อยที่สุดสำหรับค่าที่มี NaN น้อยมาก
df["Smoked 100+ Cigarettes"].fillna(df["Smoked 100+ Cigarettes"].mode()[0], inplace=True)
df["Doctor Visit Cost Barrier"].fillna(df["Doctor Visit Cost Barrier"].mode()[0], inplace=True)

# บันทึกข้อมูลที่ทำ Feature Engineering แล้ว
final_processed_path = "/kaggle/working/processed_train.csv"
df.to_csv(final_processed_path, index=False)

print(f"Final processed dataset saved at: {final_processed_path}")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score
import numpy as np

def preprocess_data(df, is_train=True):
    """ทำ Feature Engineering และ Preprocessing ข้อมูล"""
    df = df.copy()
    if is_train:
        df = df.dropna(subset=["History of HeartDisease or Attack"])
    
    num_imputer = SimpleImputer(strategy="median")
    df.loc[:, "Body Mass Index"] = num_imputer.fit_transform(df[["Body Mass Index"]])
    
    categorical_cols = ["Told High Cholesterol", "General Health", "Diagnosed Diabetes", "Difficulty Walking"]
    categorical_imputer = SimpleImputer(strategy="most_frequent")
    df.loc[:, categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])
    
    binary_cols = [
        "High Blood Pressure", "Told High Cholesterol", "Cholesterol Checked",
        "Smoked 100+ Cigarettes", "Diagnosed Stroke", "Diagnosed Diabetes",
        "Leisure Physical Activity", "Heavy Alcohol Consumption",
        "Health Care Coverage", "Doctor Visit Cost Barrier", "Difficulty Walking",
        "Vegetable or Fruit Intake (1+ per Day)"
    ]
    for col in binary_cols:
        df.loc[:, col] = df[col].map({"Yes": 1, "No": 0})
    
    ordinal_mappings = {
        "General Health": {"Very Poor": 1, "Poor": 2, "Fair": 3, "Good": 4, "Excellent": 5},
        "Education Level": {"Never attended school": 1, "Some school": 2, "High school graduate": 3, 
                            "Some college": 4, "College graduate": 5},
        "Income Level": {"Less than $10,000": 1, "$10,000 - $14,999": 2, "$15,000 - $24,999": 3, 
                         "$25,000 - $34,999": 4, "$35,000 - $49,999": 5, "$50,000 - $74,999": 6, 
                         "$75,000 or more": 7}
    }
    for col, mapping in ordinal_mappings.items():
        df.loc[:, col] = df[col].map(mapping)
    
    df.loc[:, "Sex"] = df["Sex"].map({"Male": 1, "Female": 0})
    df.loc[:, list(ordinal_mappings.keys())] = num_imputer.fit_transform(df[list(ordinal_mappings.keys())])
    df = df.drop(columns=["ID"], errors="ignore")
    
    return df

def train_logistic_regression(X_train, y_train, X_val, y_val):
    """Logistic Regression ใช้ SMOTE เพื่อจัดการปัญหา Class Imbalance"""
    smote = SMOTE(random_state=42)
    X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)
    
    model = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)
    model.fit(X_train_resampled, y_train_resampled)
    
    y_val_pred = model.predict(X_val)
    accuracy = accuracy_score(y_val, y_val_pred)
    print(f"Logistic Regression Accuracy: {accuracy:.4f}")
    
    return model

def predict_and_submit(best_model, X_test, test_ids):
    """submission file"""
    y_test_pred = best_model.predict(X_test)
    y_test_pred = np.where(y_test_pred == 1, "Yes", "No")
    
    submission_df = pd.DataFrame({"ID": test_ids.flatten(), "History of HeartDisease or Attack": y_test_pred})
    submission_output_path = "/kaggle/working/submission.csv"
    submission_df.to_csv(submission_output_path, index=False)
    print(f"Submission file saved at: {submission_output_path}")

# โหลดข้อมูล
train_path = "/kaggle/input/hearth-disease-recognition/train.csv"
test_path = "/kaggle/input/hearth-disease-recognition/test.csv"

df_train = pd.read_csv(train_path)
df_test = pd.read_csv(test_path)

# ทำ Feature Engineering
train_df_clean = preprocess_data(df_train, is_train=True)
test_df_clean = preprocess_data(df_test, is_train=False)
test_ids = df_test[["ID"]].values

# แยก X และ y
X = train_df_clean.drop(columns=["History of HeartDisease or Attack"])
y = train_df_clean["History of HeartDisease or Attack"].map({"Yes": 1, "No": 0})

# Normalization
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
test_scaled = scaler.transform(test_df_clean)

# เติมค่า NaN
imputer = SimpleImputer(strategy="median")
X_scaled = imputer.fit_transform(X_scaled)
test_scaled = imputer.transform(test_scaled)

# แยก train และ validation set
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# ฝึก Logistic Regression พร้อม SMOTE
best_model = train_logistic_regression(X_train, y_train, X_val, y_val)

# ทำนายและสร้าง submission file
predict_and_submit(best_model, test_scaled, test_ids)
```
